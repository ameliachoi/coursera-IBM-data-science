### Line Plots (Series/Dataframe)

Q. Plot a line graph of immigration from Haiti using `df.plot()`

```
haiti = df_can.loc['Haiti',years]
haiti.head()
```

```
haiti.plot()
```

```
# label the x and y axis
haiti.index = haiti.index.map(int) # change the index values of haiti to type integer for plotting
haiti.plot(kind='line') # line plot

plt.title('Immigration from Haiti')
plt.xlabel('Years')
plt.ylabel('Number of immigrants')

plt.show()
```

```
# annotate this spike in the plot
haiti.plot(kind='line')

plt.title('Immigration from Haiti')
plt.xlabel('Years')
plt.ylabel('Number of immigrants')

plt.text(2000, 6000, '2010 Earthquake')
plt.show()
```

Q. Let's compare the number of immigrants from India and China from 1980 to 2013.

```
# get the dataset for China and India, and display dataframe
df_CI = df_can.loc[['India','China'], years]
df_CI.head()
```

```
# plot graph
df_CI.plot(kind='line')
```

```
# it doesn't look right...
# must first transpose the dataframe to swap the row and columns
df_CI = df_CI.transpose()
df_CI.head()
```

```
df_CI.index = df_CI.index.map(int)
df_CI.plot(kind='line')

plt.title('The number of immigrants from China and India')
plt.xlabel('Years')
plt.ylabel('Number of immigrants')

plt.show()
```

Q. Compare the trend of top 5 countries that contributed the most to immigrant to Canada

```
df_top5 = df_can.head(5)
df_top5 = df_top5[years].transpose()
print(df_top5)
```

```
df_top5.index = df_top5.index.map(int)
df_top5.plot(kind='line')

plt.title('Immigration Trend of Top 5 Countries')
plt.xlabel('Years')
plt.ylabel('Number of Immigrants')

plt.show()
```
